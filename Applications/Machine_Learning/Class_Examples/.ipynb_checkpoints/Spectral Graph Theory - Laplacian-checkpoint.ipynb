```python
import networkx as nx
import numpy as np
import scipy
from scipy import linalg
import matplotlib.pyplot as plt
```

Path Graph

```python
G = nx.path_graph(10)

pos = nx.spring_layout(G)
nx.draw(G, pos=pos)
```

```python
L = nx.laplacian_matrix(G)

L.todense()
```

```python
eigvals, eigvecs = linalg.eigh(L.todense())
```

```python
eigvals
```

```python
eigvecs[:,0]
```

```python
eigvecs[:,1]
```

```python
colors = eigvecs[:,1]
vmin = colors.min()
vmax = colors.max()
cmap = plt.cm.coolwarm

nx.draw_networkx(G, pos=pos, node_color=colors,
                 cmap=cmap, vmin=vmin, vmax=vmax, with_labels=False)

sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm.set_array([])
cbar = plt.colorbar(sm)
```

```python
colors = eigvecs[:,2]
vmin = colors.min()
vmax = colors.max()
cmap = plt.cm.coolwarm

nx.draw_networkx(G, pos=pos, node_color=colors,
                 cmap=cmap, vmin=vmin, vmax=vmax, with_labels=False)

sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm.set_array([])
cbar = plt.colorbar(sm)
```

```python
colors = eigvecs[:,3]
vmin = colors.min()
vmax = colors.max()
cmap = plt.cm.coolwarm

nx.draw_networkx(G, pos=pos, node_color=colors,
                 cmap=cmap, vmin=vmin, vmax=vmax, with_labels=False)

sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm.set_array([])
cbar = plt.colorbar(sm)
```

```python
colors = eigvecs[:,-1]
vmin = colors.min()
vmax = colors.max()
cmap = plt.cm.coolwarm

nx.draw_networkx(G, pos=pos, node_color=colors,
                 cmap=cmap, vmin=vmin, vmax=vmax, with_labels=False)

sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm.set_array([])
cbar = plt.colorbar(sm)
```
